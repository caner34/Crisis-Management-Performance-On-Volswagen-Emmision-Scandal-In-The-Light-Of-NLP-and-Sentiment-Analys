```python
import tweepy
import json
from time import time
```

```python

credentials = {}  
credentials['CONSUMER_KEY'] = "" 
credentials['CONSUMER_SECRET'] = ""
credentials['ACCESS_TOKEN'] = ""
credentials['ACCESS_TOKEN_SECRET'] = ""

f = open('twitter_credentials_from_web.json', 'r')
credential_data = f.read()

credentials = json.loads(credential_data)

    
auth = tweepy.OAuthHandler(credentials['CONSUMER_KEY'], credentials['CONSUMER_SECRET'])
auth.set_access_token(credentials['ACCESS_TOKEN'], credentials['ACCESS_TOKEN_SECRET'])



# Create API object
api = tweepy.API(auth)

try:
    api.verify_credentials()
    print("API created")
except Exception as e:
    print("Error creating API: "+str(e))

def print_twitter_rate_limits():
    data = api.rate_limit_status()
    print(data)
    print("\n\n")
    print(data['resources']['statuses']['/statuses/home_timeline'])
    print("\n\n")
    print(data['resources']['users']['/users/lookup'])


```

```python
import GetOldTweets3 as got
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
def get_tweets_from_user(user, max_tweets):
    username = user
    count = max_tweets
    
    # Creation of query object
    tweetCriteria = got.manager.TweetCriteria().setUsername(username)\
                                            .setMaxTweets(count)
    # Creation of list that contains all tweets
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)# Creating list of chosen tweet data
    user_tweets = [[tweet.date, tweet.text] for tweet in tweets]
    
    return tweets, user_tweets
```

```python
def get_tweets_from_search(query_term, max_tweets):
    text_query = query_term
    count = max_tweets
    
    # Creation of query object
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query).setMaxTweets(count)
    # Creation of list that contains all tweets
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    # Creating list of chosen tweet data
    text_tweets = [[tweet.date, tweet.text] for tweet in tweets]
    
    return tweets, text_tweets
```

```python
def get_tweets_from_search_by_date(query_term, max_tweets, since_date_YYYY_dash_MM_dash_DD, until_date_YYYY_dash_MM_dash_DD):
    text_query = query_term
    since_date = since_date_YYYY_dash_MM_dash_DD
    until_date = until_date_YYYY_dash_MM_dash_DD
    count = max_tweets
    
    # Creation of query object
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(text_query)\
    .setSince(since_date).setUntil(until_date).setMaxTweets(count)
    
    # Creation of list that contains all tweets
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    # Creating list of chosen tweet data
    text_tweets = [[tweet.date, tweet.text] for tweet in tweets]
    
    return tweets, text_tweets
```

```python
negative_comments = 0
total_comments = 0

analyzer = SentimentIntensityAnalyzer()

for t in user_tweets_emission_15_09_20:
    cr_sentiment_neutral = analyzer.polarity_scores(t[1])['neu']
    cr_sentiment_compound = analyzer.polarity_scores(t[1])['compound']
    print("* "+t[1]+"\n"+str(t[0])+"\n"+str(cr_sentiment_compound)+"\n"+str(cr_sentiment_neutral)+"\n")
    total_comments += 1
    if cr_sentiment_compound < 0.0:
        negative_comments += 1

print("negative_ratio: {}".format((negative_comments/total_comments)))
```
